```python
#task 2.a
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('cameraman.tif')

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Image loaded using OpenCV")
plt.axis('off')
plt.show()
```

```python
#task 2.b

from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.cm as cm

img2 = Image.open('bark_he_512.tiff')

plt.imshow(img2, cmap=cm.Greys_r)
plt.title("Image loaded using PIL")
plt.axis('off')
plt.show()
```
