```python
import cv2

# Load image in grayscale
img = cv2.imread("pngtree-a-cute-parrot-picture-png-image_14102326.png", cv2.IMREAD_GRAYSCALE)

# Threshold values
threshold_values = [0, 50, 100, 150, 200]

# Apply thresholding
for t in threshold_values:
    ret, thresh = cv2.threshold(img, t, 255, cv2.THRESH_BINARY)
    cv2.imshow(f'Threshold {t}', thresh)

# Show original
cv2.imshow('Original Image', img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import data
from skimage import exposure

# Load moon image
img = data.moon()

# Contrast Stretching (2% and 98%)
p2, p98 = np.percentile(img, (2, 98))

img_rescale = exposure.rescale_intensity(img, in_range=(p2, p98))

# Display original image + histogram
plt.figure(figsize=(10,7))

plt.subplot(2,1,1)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title('Original Image')

plt.subplot(2,1,2)
plt.hist(img.flat, bins=256, range=(0,255))
plt.xlabel('Gray Level')
plt.ylabel('Pixels')

plt.show()

# Display stretched image + histogram
plt.figure(figsize=(10,7))

plt.subplot(2,1,1)
plt.imshow(img_rescale, cmap='gray')
plt.axis('off')
plt.title('Contrast Stretching')

plt.subplot(2,1,2)
plt.hist(img_rescale.flat, bins=256, range=(0,255))
plt.xlabel('Gray Level')
plt.ylabel('Pixels')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, exposure

img = data.moon()

# 3% and 80%
p3, p80 = np.percentile(img, (3, 80))

img_rescale = exposure.rescale_intensity(img, in_range=(p3, p80))

plt.figure(figsize=(10,7))

plt.subplot(2,1,1)
plt.imshow(img_rescale, cmap='gray')
plt.title("Rescaled Image (3%,80%)")
plt.axis('off')

plt.subplot(2,1,2)
plt.hist(img_rescale.flat, bins=256, range=(0,255))
plt.xlabel('Gray Level')
plt.ylabel('Pixels')

plt.show()
```

```python
from skimage import data, exposure
import matplotlib.pyplot as plt

img = data.moon()

# Equalize histogram
img_eq = exposure.equalize_hist(img)

plt.figure(figsize=(10,7))

plt.subplot(2,1,1)
plt.imshow(img_eq, cmap='gray')
plt.title("Equalized Image")
plt.axis('off')

plt.subplot(2,1,2)
plt.hist(img_eq.flat, bins=256)
plt.xlabel('Gray Level')
plt.ylabel('Pixels')

plt.show()
```

```python
from skimage import data, exposure
import matplotlib.pyplot as plt

# Load images
reference = data.rocket()
image = data.chelsea()

# Histogram matching
matched = exposure.match_histograms(image, reference, channel_axis=-1)

# Show results
plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.imshow(image)
plt.title("Chelsea (Original)")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(reference)
plt.title("Rocket (Reference)")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(matched)
plt.title("Matched Image")
plt.axis('off')

plt.show()
```
